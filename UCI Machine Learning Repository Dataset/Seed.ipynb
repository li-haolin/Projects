```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seed = pd.read_csv("/Users/jeremyli/Desktop/STA160Template/seeds_dataset.txt", header = None, names = ["data","area","perimeter","compactness","length", "width","asymmetry","lengthOfGroove","varieties"] ) 
seed
```

```python
for i in range(210):
    sep = seed["data"][i].split("\t")
    seed["data"][i] = sep
seed
```

```python
for i in range(210):
    if len(seed["data"][i]) > 8:
        while("" in seed["data"][i]):
            seed["data"][i].remove("")

seed
```

```python
for i in range(210):
    seed["area"][i] = seed["data"][i][0]
    seed["perimeter"][i] = seed["data"][i][1]
    seed["compactness"][i] = seed["data"][i][2]
    seed["length"][i] = seed["data"][i][3]
    seed["width"][i] = seed["data"][i][4]
    seed["asymmetry"][i] = seed["data"][i][5]
    seed["lengthOfGroove"][i] = seed["data"][i][6]
    seed["varieties"][i] = seed["data"][i][7]
seed 
```

```python
seed.head()
```

```python
kama = seed[seed.varieties == 1]
rosa = seed[seed.varieties == 2]
cana = seed[seed.varieties == 3]
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=15)

plt.hist(kama["area"], **kwargs, label='Kama')
plt.hist(rosa["area"], **kwargs, label='Rosa')
plt.hist(cana["area"], **kwargs, label='Canadian')

plt.xlabel("Area")
plt.title("Histogram of Area of 3 Wheats")
plt.legend()
```

```python


plt.hist(kama["perimeter"], **kwargs, label='Kama')
plt.hist(rosa["perimeter"], **kwargs, label='Rosa')
plt.hist(cana["perimeter"], **kwargs, label='Canadian')
plt.xlabel("Perimeter")
plt.title("Histogram of Perimeter of 3 Wheats")
plt.legend();
```

```python


plt.hist(kama["compactness"], **kwargs, label='Kama')
plt.hist(rosa["compactness"], **kwargs, label='Rosa')
plt.hist(cana["compactness"], **kwargs, label='Canadian')
plt.xlabel("Compactness")
plt.title("Histogram of Compactness of 3 Wheats")
plt.legend();
```

```python


plt.hist(kama["length"], **kwargs, label='Kama')
plt.hist(rosa["length"], **kwargs, label='Rosa')
plt.hist(cana["length"], **kwargs, label='Canadian')
plt.xlabel("Length of Kernel")
plt.title("Histogram of Length of 3 Wheats")
plt.legend();
```

```python
plt.hist(kama["width"], **kwargs, label='Kama')
plt.hist(rosa["width"], **kwargs, label='Rosa')
plt.hist(cana["width"], **kwargs, label='Canadian')

plt.xlabel("Width of Kernel")
plt.title("Histogram of Width of 3 Wheats")
plt.legend()

```

```python
plt.hist(kama["asymmetry"], **kwargs, label='Kama')
plt.hist(rosa["asymmetry"], **kwargs, label='Rosa')
plt.hist(cana["asymmetry"], **kwargs, label='Canadian')

plt.xlabel("Asymmetry")
plt.title("Histogram of Asymmetry of 3 Wheats")
plt.legend()

```

```python
plt.hist(kama["lengthOfGroove"], **kwargs, label='Kama')
plt.hist(rosa["lengthOfGroove"], **kwargs, label='Rosa')
plt.hist(cana["lengthOfGroove"], **kwargs, label='Canadian')

plt.xlabel("Length of Kernel Groove")
plt.title("Histogram of Length of Kernel Groove of 3 Wheats")
plt.legend()

```

```python
import seaborn as sns
```

```python
seed_x = [seed["area"],seed["perimeter"],seed["compactness"],seed["length"],seed["width"],seed["asymmetry"],seed["lengthOfGroove"]]
ax = sns.clustermap(seed_x, cmap="plasma", standard_scale=1)
ax.fig.suptitle("Cluster Map of Features of Wheats")
```

```python

from matplotlib import cm

feature_names = ['area', 'perimeter', 'compactness', 'length','width','asymmetry','lengthOfGroove']
X = seed[feature_names]
y = seed['varieties']
cmap = cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(X, c = y, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(9,9), cmap = cmap)
plt.suptitle('Scatter-matrix for Each Feature')
```

```python
plt.figure(figsize=[8,8])
sns.heatmap(seed.corr(), annot=True, cmap="coolwarm")
plt.title('Correlations of the Features')
plt.show()
```

```python
df = seed.iloc[:,0:7]
df = df.drop("data",axis=1)
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
plt.figure(figsize=[10,10])
merg = linkage(df, method='ward')
dendrogram(merg, leaf_rotation=90)
plt.title('Dendrogram of Observations')
plt.xlabel('Observations')
plt.ylabel('Distance')
plt.show()
```
