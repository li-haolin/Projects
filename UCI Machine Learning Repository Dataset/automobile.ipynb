```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
auto = pd.read_table("/Users/jeremyli/Desktop/STA160Template/imports-85.data",header = None,
                    names = ["data","symboling","n_losses","make","fuel_type", "aspiration","doors","body_style","drive_wheels"
                            ,"engine_location","wheels","length","width","height","weight","engine_type","cylinders"
                            ,"engine_size","fuel_system","bore","stroke","compression","horsepower","peak","city"
                            ,"highway","price"])
auto
```

```python
for i in range(205):
    sep = auto["data"][i].split(",")
    auto["data"][i] = sep
auto
```

```python
for i in range(205):
    auto["symboling"][i] = auto["data"][i][0]
    auto["n_losses"][i] = auto["data"][i][1]
    auto["make"][i] = auto["data"][i][2]
    auto["fuel_type"][i] = auto["data"][i][3]
    auto["aspiration"][i] = auto["data"][i][4]
    auto["doors"][i] = auto["data"][i][5]
    auto["body_style"][i] = auto["data"][i][6]
    auto["drive_wheels"][i] = auto["data"][i][7]
    auto["engine_location"][i] = auto["data"][i][8]
    auto["wheels"][i] = auto["data"][i][9]
    auto["length"][i] = auto["data"][i][10]
    auto["width"][i] = auto["data"][i][11]
    auto["height"][i] = auto["data"][i][12]
    auto["weight"][i] = auto["data"][i][13]
    auto["engine_type"][i] = auto["data"][i][14]
    auto["cylinders"][i] = auto["data"][i][15]
    auto["engine_size"][i] = auto["data"][i][16]
    auto["fuel_system"][i] = auto["data"][i][17]
    auto["bore"][i] = auto["data"][i][19]
    auto["stroke"][i] = auto["data"][i][19]
    auto["compression"][i] = auto["data"][i][20]
    auto["horsepower"][i] = auto["data"][i][21]
    auto["peak"][i] = auto["data"][i][22]
    auto["city"][i] = auto["data"][i][23]
    auto["highway"][i] = auto["data"][i][24]
    auto["price"][i] = auto["data"][i][25]
    
auto
```

```python
nl = auto['n_losses'].loc[auto['n_losses'] == '?'].count()

nlmean = nl.astype(int).mean()
auto['n_losses'] = auto['n_losses'].replace('?',nlmean).astype(int)
auto['n_losses'].head()
```

```python
price = auto['price'].loc[auto['price'] != '?']
pmean = price.astype(int).mean()
auto['price'] = auto['price'].replace('?',pmean).astype(int)
auto['price'].head()
```

```python
auto['horsepower'].str.isnumeric().value_counts()
horsepower = auto['horsepower'].loc[auto['horsepower'] != '?']
hpmean = horsepower.astype(int).mean()
auto['horsepower'] = auto['horsepower'].replace('?',pmean).astype(int)
```

```python
auto['bore'] = pd.to_numeric(auto['bore'],errors='coerce')
auto['stroke'] = pd.to_numeric(auto['stroke'],errors='coerce')
auto['peak'] = pd.to_numeric(auto['peak'],errors='coerce')
```

```python
auto['doors'].loc[auto['doors'] == '?']
auto = auto[auto['doors'] != '?']
auto['doors'].loc[auto['doors'] == '?']
auto["doors"]
```

```python
import seaborn as sns
corr = auto.corr()
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(13,7))
a = sns.heatmap(corr, annot=True, fmt='.2f')
rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)
rott = a.set_title("Correlation Matrix of Automobile")
```

```python
auto['body_style'].value_counts().plot.pie(figsize=(6, 6), autopct='%.2f')
plt.title("Body Style Pie Diagram")
plt.ylabel('Number of vehicles')
plt.xlabel('Body Style');
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(auto.engine_size[auto.fuel_type=="gas"], **kwargs, label='gas')
plt.hist(auto.engine_size[auto.fuel_type=="diesel"], **kwargs, label='diesel')
plt.xlabel("Engine Size")
plt.title("Histogram of Engine Size in Gas and Diesel")
plt.legend();
```
